# **Detección de Colinealidad**
Identificar variables altamente correlacionadas

```python
# Identificar variables altamente correlacionadas

# Calcular la matriz de correlación
correlation_matrix = data.corr()

# Establecer un umbral para correlaciones altas
threshold = 0.8

# Encontrar variables altamente correlacionadas
highly_correlated_vars = correlation_matrix[abs(correlation_matrix) > threshold]

# Imprimir las variables
highly_correlated_vars
```

**Variables altamente correlacionadas:**

- App usage time, Screen on time, Battery drain, Number of apps, Data Usage, User behavior class.



```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Selección de las columnas numéricas relevantes (sin incluir la variable objetivo)
features = [
    "App Usage Time (min/day)",
    "Screen On Time (hours/day)",
    "Battery Drain (mAh/day)",
    "Number of Apps Installed",
    "Data Usage (MB/day)",
    "Age"
]

# 1. Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[features])

# 2. Aplicar PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 3. Explicar la varianza
explained_variance = pca.explained_variance_ratio_

# Mostrar la varianza acumulada
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance)+1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.title('Varianza acumulada explicada por los componentes principales')
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza acumulada')
plt.grid()
plt.show()

# 4. Imprimir resultados
print("Varianza explicada por cada componente principal:")
for i, var in enumerate(explained_variance, 1):
    print(f"Componente {i}: {var:.4f}")

print("\nVarianza acumulada con 2 componentes principales:")
print(f"{explained_variance[:2].sum():.4f}")
```

Por tanto elegimos la componente 1 y 2: "App Usage Time (min/day)", "Screen On Time (hours/day)" y eliminamos el resto de las variables.



```python
# Transformar los datos usando solo los 2 componentes principales
X_reduced = pca.transform(X_scaled)[:, :2]

# Crear un nuevo DataFrame con los componentes principales
df_pca = pd.DataFrame(X_reduced, columns=['PCA1', 'PCA2'])

# Si deseas incluir la variable objetivo:
df_pca['User Behavior Class'] = data['User Behavior Class']

# Mostrar las primeras filas del DataFrame reducido
print(df_pca.head())
```

# **División de conjunto de Datos y Aplicación del Modelo**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Crear el DataFrame reducido
df_pca = pd.DataFrame(X_reduced, columns=['PCA1', 'PCA2'])
df_pca['User Behavior Class'] = data['User Behavior Class']

# Seleccionar las variables reducidas para el pipeline
X = df_pca[['PCA1', 'PCA2']]
y = df_pca['User Behavior Class']

# 1. Normalizar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Dividir en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# 3. Dividir el conjunto de entrenamiento en entrenamiento (70%) y validación (30%)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)

# 4. Crear y entrenar un modelo de Random Forest (multiclase)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 5. Evaluar el modelo en el conjunto de validación
y_val_pred = model.predict(X_val)
print("Evaluación en el conjunto de validación:")
print("Accuracy (Validación):", accuracy_score(y_val, y_val_pred))
print("\nClassification Report (Validación):")
print(classification_report(y_val, y_val_pred))

# 6. Realizar predicciones en el conjunto de prueba
y_test_pred = model.predict(X_test)

# 7. Evaluar el modelo en el conjunto de prueba
print("\nEvaluación en el conjunto de prueba:")
print("Accuracy (Prueba):", accuracy_score(y_test, y_test_pred))
print("\nConfusion Matrix (Prueba):")
print(confusion_matrix(y_test, y_test_pred))
print("\nClassification Report (Prueba):")
print(classification_report(y_test, y_test_pred))

# Opcional: Visualizar la Confusion Matrix para el conjunto de prueba
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=model.classes_, yticklabels=model.classes_)
plt.title('Matriz de Confusión - Conjunto de Prueba')
plt.xlabel('Predicciones')
plt.ylabel('Verdaderas Clases')
plt.show()
```

# **División de conjunto de Datos y Aplicación del Modelo con Validación cruzada**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Crear el DataFrame reducido
df_pca = pd.DataFrame(X_reduced, columns=['PCA1', 'PCA2'])
df_pca['User Behavior Class'] = data['User Behavior Class']

# Seleccionar las variables reducidas para el pipeline
X = df_pca[['PCA1', 'PCA2']]
y = df_pca['User Behavior Class']

# 1. Normalizar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Crear el modelo
model = RandomForestClassifier(random_state=42)

# 3. Definir validación cruzada (K-Fold)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)  # 5-fold cross-validation

# 4. Evaluar el modelo usando cross_val_score
scores = cross_val_score(model, X_scaled, y, cv=kfold, scoring='accuracy')
print("Accuracy por fold:", scores)
print("Promedio de Accuracy:", scores.mean())

# 5. Predicciones con validación cruzada
y_pred = cross_val_predict(model, X_scaled, y, cv=kfold)

# 6. Reportar métricas
print("\nClassification Report:")
print(classification_report(y, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y, y_pred))
```

# **Aplicación de PCA en cada fold de la validación cruzada para evitar data leakage.**

Asegurarnos de que no haya información del conjunto de datos de prueba, validación o de los folds que se esté filtrando en el entrenamiento.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report

# Transformar los datos usando PCA previamente
X_reduced = pca.transform(X_scaled)[:, :2]  # Usar solo las dos primeras componentes principales

# Crear el DataFrame reducido
df_pca = pd.DataFrame(X_reduced, columns=['PCA1', 'PCA2'])
df_pca['User Behavior Class'] = data['User Behavior Class']

# Seleccionar las variables reducidas para el pipeline
X = df_pca[['PCA1', 'PCA2']]
y = df_pca['User Behavior Class']

# Configurar el pipeline con normalización y modelo
pipeline = Pipeline([
    ('scaler', StandardScaler()),          # Normalización
    ('model', RandomForestClassifier(random_state=42))  # Modelo Random Forest
])

# Configurar validación cruzada estratificada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Calcular el desempeño con validación cruzada
scores = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')

# Resultados
print("Resultados de validación cruzada (accuracy):", scores)
print("Promedio de accuracy:", scores.mean())

# (Opcional) Entrenar y predecir con todo el conjunto reducido
pipeline.fit(X, y)
y_pred = pipeline.predict(X)

# Evaluar predicciones con matriz de confusión
print("\nMatriz de confusión:")
print(confusion_matrix(y, y_pred))
print("\nReporte de clasificación:")
print(classification_report(y, y_pred))
```
